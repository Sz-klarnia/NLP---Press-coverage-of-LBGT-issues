```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
# using selenium to scroll the page
driver = webdriver.Firefox()
driver.get("https://klubjagiellonski.pl/?s=lgbt")
last_height = driver.execute_script("return document.body.scrollHeight")
for i in range(30):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")

```

```python
filename = "kj.csv"
f=open(filename,"w",encoding="utf-8")
headers="Title,Content,Link,Date,Author\n"
f.write(headers)


soup = BeautifulSoup(driver.page_source, "html.parser")
links = soup.find_all("div",attrs={"class":"block-postlink_featured block-postlink postlink_post_type-post"})

for link in links:
    date = link.find("span",attrs={"class":"data"}).text
    if date[-4:] == "2020":
        
        title = link.find("h1").text
        url = link.find("a",attrs={"class":"in"}).get("href").strip()
        print(f"page processed: {url}")
        request = requests.get(url)
        soup1 = BeautifulSoup(request.text, "html.parser")
        author = soup1.find("p",attrs={"class":"imienazwisko"}).text
        content_list = soup1.find_all("p")
        content = ""
        for p in content_list:
            content = content + p.text
        content = content.strip().replace("\n","")
        content = content.replace("\xa0"," ")
        f.write(title.replace(",","^")+","+content.replace(",","^")+","+url+","+date.replace(",","^")+","+author.replace(",","^")+"\n")
        
f.close()
```
