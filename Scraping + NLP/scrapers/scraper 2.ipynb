```python
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
date = "2020"

filename="niezalezna.csv"
f=open(filename,"w", encoding = 'utf-8')
headers="Title,Content,Link,Date,Author\n"
f.write(headers)
loop_nr = 0
page_count=1

while date=="2020":
    url = f"https://niezalezna.pl/wyszukiwarka/{page_count}?s=lgbt"
    print(loop_nr)
    print(url)
    page = requests.get(url)
    time.sleep(2)
    soup = BeautifulSoup(page.text, "html.parser")
    links = soup.find_all("div",attrs={"class":"articleItem articleItemHorizontal articleItemHorizontalH150 boxShadow"})
    for link in links:
        url1 = "https://niezalezna.pl/" + (link.find("a").get("href").strip())
        print(f"Link processed {url1}")
        page1 = requests.get(url1)
        time.sleep(2)
        soup1 = BeautifulSoup(page1.text, "html.parser")
        content_list = soup1.find("p").find_all("p")
        content = ""
        for p in content_list:
            content = content + p.text
        content = content.strip().replace("\n","")
        try:
            title = soup1.find("div",attrs={"class":"articleContent"}).find("h1").text.strip()
        except:
            title = soup1.find("article",attrs={"class":"articleContent"}).find("h1").text.strip()
        try:
            date_internal = soup1.find("div",attrs={"class":"articleTypeAndTime"}).find("span").text[0:10]
        except:
            date_internal = soup1.find("span",attrs={"class":"newsTime"}).text[0:10]
        try:
            author = soup1.find("div",attrs={"class":"authorName"}).find("a").text
        except:
            try:
                author = soup1.find("p",attrs={"class":"authorName"}).find("a").text
            except:
                author = "brak"
        f.write(title.replace(",","^")+","+content.replace(",","^")+","+url1+","+date_internal.replace(",","^")+","+author.replace(",","^")+"\n")
    date = date_internal[6:]
    print(date_internal)
    loop_nr += 1
    page_count+=1
f.close()
```
