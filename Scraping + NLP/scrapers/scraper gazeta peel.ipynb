```python
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
filename = "gazetapeel.csv"
f=open(filename,"w",encoding="utf-8")
headers="Title,Content,Link,Date,Author\n"
f.write(headers)
date = "2020"
loop_nr = 1
page_number = 1
while date == "2020":
    print(f"page number {page_number}")
    url =f"https://szukaj.gazeta.pl/wyszukaj/artykul?&query=lgbt&navservice=Wiadomo%B6ci&sortMode=DATE&pageNumber={page_number}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text,"html.parser")
    links = soup.find_all("section",attrs={"class":"elem normal"})
    for link in links:
        
        date_internal = link.find("time").text
        author = link.find("span",attrs={"class":"author"}).text
        try:
            url1 = link.find("a",attrs={"class":""}).get("href").strip()
        except:
            url1 = link.find("a",attrs={"class":"restriction_0"}).get("href").strip()

        print(f"Link processed {url1}")
        try:
            title = link.find("a",attrs={"class":""}).text
        except:
            title = link.find("a",attrs={"class":"restriction_0"}).text
        page1 = requests.get(url1)
        soup1 = BeautifulSoup(page1.text,"html.parser")
        content_list = soup1.find_all("p")
        content = ""
        for p in content_list:
            content = content + p.text
        content = content.strip().replace("\n","")
        content = content.replace("\xa0"," ")
        f.write(title.replace(",","^")+","+content.replace(",","^")+","+url1.replace(",",".")+","+date_internal.replace(",","^")+","+author.replace(",","^")+"\n")
    loop_nr += 1
    page_number +=1 
    date = date_internal[6:10]
f.close()
```
