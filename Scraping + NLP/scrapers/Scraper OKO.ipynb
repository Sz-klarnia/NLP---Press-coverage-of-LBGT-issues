```python
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python

filename="oko.csv"
f=open(filename,"w", encoding = 'utf-8')
headers="Title,Content,Link,Date,Author\n"
f.write(headers)

for i in range(0,26):
    page_count = i+1
    url = f"https://www.pch24.pl/informacje/szukajz?q1=lgbt&dod=2019-12-31&ddo=2020-12-20&where=3&sortby=dp01&mpp=&strona={page_count}"
    print(page_count)
    page = requests.get(url)
    time.sleep(2)
    soup = BeautifulSoup(page.text, "html.parser")
    links1 = soup.find_all("div",attrs={"class":"TplWarto bggrey pb10 mt20"})
    links2 = soup.find_all("div",attrs={"class":"TplWarto bggrey pb10 mt20"})
    links = soup.find_all("div",attrs={"class":"TplWarto bggrey pb10 mt20"})

    for i in range(len(links)):
        url1 = links[i].find("h2").find("a").get("href").strip()
        if type(re.search(r"wypowiedzi",url1)) == re.Match:
            continue
        print(f"Link processed {url1}")
        page1 = requests.get(url1)
        time.sleep(2)
        soup1 = BeautifulSoup(page1.text, "html.parser")
        content_list = soup1.find("div",
                        attrs={"class":"entry-content"}).find_all("p")
        content = ""
        for p in content_list:
            content = content + p.text
        content = content.strip().replace("\n","") 
        date = soup1.find("time", attrs={"class":"updated"}).text
        title = links[i].find("h2").find("a").text
        try:
            author = soup1.find("span",attrs={"class":"meta-section__autor"}).text.strip()
        except:
            author = soup1.find("a",attrs={"class":"entry-meta__author"}).text
        if date[-4:]=="2020":
            f.write(title.replace(",","^")+","+content.replace(",","^")+","+url1+","+date.replace(",","^")+","+author.replace(",","^")+"\n")

    loop_nr += 1
f.close()
```
