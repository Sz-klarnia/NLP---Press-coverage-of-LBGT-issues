```python

import time
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
filename = "se.csv"
f=open(filename,"w",encoding="utf-8")
headers="Title,Content,Link,Date,Author\n"
f.write(headers)
page_nr = 1
for i in range(30):
    print(page_nr)
    url =f"https://www.se.pl/temat/lgbt,tg-NtRa-29x2-WoiF/?page={page_nr}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text,"html.parser")
    links = soup.find_all("div",attrs={"class":"element"})
    for link in links:
        url1 = "http:"+link.find("a").get("href").strip()
        if type(re.search(r"se.pl",url1)) == re.Match:
            print(f"processed link: {url1}")
            page1 = requests.get(url1)
            soup1 = BeautifulSoup(page1.text,"html.parser")
            date = soup1.find("span",attrs={"class":"h3 pub_time_date"}).text
            author = soup1.find("span",attrs={"class":"h3"}).text
            title = soup1.find("div", attrs={"class":"title"}).text.replace("\n","")
            content_list = soup1.find_all("p")
            content = ""
            for p in content_list:
                content = content + p.text
            content = content.strip().replace("\n","")
            content = content.replace("\xa0"," ")
            if date[0:4] == "2020":
                f.write(title.replace(",","^")+","+content.replace(",","^")+","+url1.replace(",",".")+","+date.replace(",","^")+","+author.replace(",","^")+"\n")
    page_nr +=1
f.close()
```
