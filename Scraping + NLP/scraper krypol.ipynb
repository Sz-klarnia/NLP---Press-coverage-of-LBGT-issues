```python
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
# creating file, writing headers
filename = "krypol.csv"
f=open(filename,"w",encoding="utf-8")
headers="Title,Content,Link,Date,Author\n"
f.write(headers)

# creating date and progress tracking variables
date = "2020"
loop_nr = 1
page_nr = ""

# looping through search pages until year changes
while date == "2020":
    
    # searchpage url
    url =f"https://krytykapolityczna.pl/temat/lgbt{page_nr}"
    # requesting and parsing page code
    page = requests.get(url)
    soup = BeautifulSoup(page.text,"html.parser")
    
    # finding article links
    links = soup.find_all("article",attrs={"itemtype":"http://schema.org/Article"})
    # iterating over articles
    for link in links:
        # saving date, author
        date_internal = link.find("time",attrs={"class":"published"}).text
        author = link.find("span",attrs={"itemprop":"name"}).text
        # article page url
        url1 = link.find("a",attrs={"itemprop":"mainEntityOfPage"}).get("href").strip()
        # saving title
        title = link.find("a",attrs={"itemprop":"mainEntityOfPage"}).text
        # progress tracking
        print(f"Link processed {url1}")
        # requesting and parsing article page code
        page1 = requests.get(url1)
        soup1 = BeautifulSoup(page1.text,"html.parser")
        
        # finding and saving article text
        content_list = soup1.find_all("p")
        content = ""
        for p in content_list:
            content = content + p.text
        content = content.strip().replace("\n","")
        content = content.replace("\xa0"," ")
        # saving data
        f.write(title.replace(",","^")+","+content.replace(",","^")+","+url1.replace(",",".")+","+date_internal.replace(",","^")+","+author.replace(",","^")+"\n")
    # updating variables
    loop_nr += 1
    date = date_internal[-4:]
    page_nr = f"/page/{loop_nr}"
    
# closing file
f.close()
```
