```python
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests

import pandas as pd
import re
```

```python
# creating file, writing headers
filename="pch241.csv"
f=open(filename,"w", encoding = 'utf-8')
headers="Title,Content,Link,Date,Author\n"
f.write(headers)

# iterating over searchpages
for i in range(0,26):
    # progress tracking
    page_count = i+1
    # page url
    url = f"https://www.pch24.pl/informacje/szukajz?q1=lgbt&dod=2019-12-31&ddo=2020-12-20&where=3&sortby=dp01&mpp=&strona={page_count}"
    print(page_count)
    
    # requesting page code
    page = requests.get(url)
    time.sleep(2)
    # parsing code
    soup = BeautifulSoup(page.text, "html.parser")
    # searching for article links (2 different types)
    links1 = soup.find_all("div",attrs={"class":"TplWarto bggrey pb10 mt20"})
    links2 = soup.find_all("div",attrs={"class":"TplWarto  mt20"})
    # creating list of links
    links = links1 + links2
    
    # iterating over links
    for link in links:
        # getting article url
        url1 = "https://www.pch24.pl/"+link.find("div",attrs={"class":"TagTitle"}).find("a").get("href").strip()
        # getting article title
        title = link.find("div",attrs={"class":"TagTitle"}).find("a").text
        # progress tracking
        print(f"Link processed {url1}")
        # requesting and parsing article code
        page1 = requests.get(url1)
        page1.encoding = "utf8"
        time.sleep(2)
        soup1 = BeautifulSoup(page1.text, "html.parser")
        
        # looking for content
        content_list = soup1.find_all("p")     
        content = ""
        # saving and precleaning text
        for p in content_list:
            content = content + p.text
        content = content.strip().replace("\n","")
        content = content.strip().replace("\xa0"," ")

        # saving other parameters
        date = soup1.find("div", attrs={"class":"flRight data"}).text
        author = "unknown"
        # writing data to file
        f.write(title.replace(",","^")+","+content.replace(",","^")+","+url1.replace(",",".")+","+date.replace(",","^")+","+author.replace(",","^")+"\n")

# closing file
f.close()
```
