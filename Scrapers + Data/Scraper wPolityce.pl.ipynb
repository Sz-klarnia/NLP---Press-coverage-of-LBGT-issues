```python
from bs4 import BeautifulSoup
import urllib.request,sys,time
import requests
import pandas as pd
import re
```

```python
# Scraping searchpage of wPolityce.pl, key word = LGBT


# date variable to stop scraping after reaching year 2019
date = "2020"

# declaring base url for searchpage
current_url = "https://wpolityce.pl/szukaj?q=lgbt&sort=najnowsze"
org_url = "https://wpolityce.pl/szukaj"
# creating and opening file to store data
filename="wpolityce1.csv"
f=open(filename,"w", encoding = 'utf-8')
headers="Title,Content,Link,Date,Author\n"
f.write(headers)

# informational variable to track progress
loop_nr = 0

# scraping in loop until date changes to 2019
while date=="2020":
    # progress tracking
    print(loop_nr)
    
    # requesting page
    page = requests.get(current_url)
    time.sleep(2)
    
    # parsing html to workable format
    soup = BeautifulSoup(page.text, "html.parser")
    
    # finding all article links on searchpage
    links = soup.find_all("li",attrs={"class":"publication-listed__single"})
    
    # looping through found links
    for link in links:
        # preparing article url
        url1 = link.find("a").get("href").strip()
        if type(re.search(r"http",url1)) == re.Match:
            continue
        # progress tracking
        print(f"Link processed {url1}")
        
        # requesting and parsing article page
        page1 = requests.get("http:"+url1)
        time.sleep(2)
        soup1 = BeautifulSoup(page1.text, "html.parser")
        
        # scraping article text
        content = soup1.find("div",
                            attrs={"class":"article__markdown publication-markdown article-body-print js-article-body js-article-body-print"}).find("div",
                            attrs={"class":"intext-links"}).text.replace("\xa0"," ")
        # scraping title
        title = link.find("h3").find("a").text.strip()
        # scraping date
        date_internal = link.find("li").find("time").get("title")[0:10]
        try:
            author = link.find("li",attrs={"class":"meta-info-list__item publication-listed__author"}).find("a").text
        except:
            author = "brak"
        # writing scraped data to file
        f.write(title.replace(",","^")+","+content.replace(",","^")+","+url1+","+date_internal.replace(",","^")+","+author.replace(",","^")+"\n")
    
    # preparing next search page url
    if loop_nr == 0:
        find_url = soup.find("div",attrs={"class":"pagination"}).find("a").get("href").strip()
    if loop_nr != 0:
        find_url = soup.find_all("div",attrs={"class":"pagination"})[0].find_all("a")[1].get("href").strip()
    current_url = org_url + find_url
    
    # updating date variable
    date = date_internal[0:4]
    # progress tracking
    print(date_internal)
    loop_nr += 1
    
# closing written file
f.close()
```
